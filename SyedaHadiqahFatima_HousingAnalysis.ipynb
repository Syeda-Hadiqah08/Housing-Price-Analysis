### **Part 1: Setup & Load Data:**

```python
# Import required Libraries:

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load and read Data:

from google.colab import files
uploaded = files.upload()
```

```python
# Load data to Pandas DataFrame:
house_df = pd.read_csv('Housing Raw Data.csv')

# View data:
house_df
```

### **Part 2: Explore & Clean Data:**

```python
# View Descriptive Statistics of the data:

print("Descriptive Statistics:")
house_df.describe()
```

```python
# Check for Missing Values:
null_val = house_df.isnull().sum()

print("Missing Values in Each Column:")
null_val
```

```python
# Check for Duplicates:
dup_check = house_df.duplicated().sum()

print("Duplicate Rows in Dataset:")
dup_check
```

### **Part 3: Compute Statistics:**

```python
def print_stats(col):
    print(f"\nStatistics for {col}:")
    print(f"Mean: {house_df[col].mean():.2f}")
    print(f"Median: {house_df[col].median():.2f}")
    print(f"Mode: {house_df[col].mode()[0]:.2f}")
    print(f"Variance: {house_df[col].var():.2f}")
    print(f"Std Dev: {house_df[col].std():.2f}")
    print(f"Skewness: {house_df[col].skew():.2f}")
    print(f"Kurtosis: {house_df[col].kurt():.2f}")
    print(f"Min: {house_df[col].min():.2f}")
    print(f"Max: {house_df[col].max():.2f}")

for column in ['price', 'area', 'bedrooms']:
    print_stats(column)
```

### **Part 4: Optional Visualization:**

```python
# Histogram for Area and Price of Houses:

# Set Figure Size:
plt.figure(figsize=(14, 6))

# Histogram for Area (left subplot):
plt.subplot(1, 2, 1)
sns.histplot(house_df['area'], bins=30, kde=True, line_kws={'color': 'navy', 'linewidth': 2}, color = 'skyblue')
plt.title('Distribution of House Area', fontweight = 'bold')
plt.xlabel('Area (sq. ft)', fontweight = 'bold')
plt.ylabel('Frequency', fontweight = 'bold')
plt.grid(True)

# Histogram for Price (right subplot):
plt.subplot(1, 2, 2)
sns.histplot(house_df['price'], bins=30, kde=True, line_kws={'color': 'darkred', 'linewidth': 2}, color = 'salmon')
plt.title('Distribution of House Prices', fontweight = 'bold')
plt.xlabel('Price', fontweight = 'bold')
plt.ylabel('Frequency', fontweight = 'bold')
plt.grid(True)

# Adjust layout to prevent overlap:
plt.tight_layout()
plt.show()
```

```python
# Box plots for Area and Price of Houses:

# Set figure size:
plt.figure(figsize=(14, 6))

# Box plot for Area (left subplot):
plt.subplot(1, 2, 1)
sns.boxplot(x=house_df['area'], color = 'khaki')
plt.title('Box Plot of Area', fontweight = 'bold')
plt.xlabel('Area (sq. ft)', fontweight = 'bold')
plt.grid(True)

# Box plot for Price (right subplot):
plt.subplot(1, 2, 2)
sns.boxplot(x=house_df['price'], color = 'teal')
plt.title('Box Plot of Price', fontweight = 'bold')
plt.xlabel('Price', fontweight = 'bold')
plt.grid(True)

# Adjust layout to prevent overlap:
plt.tight_layout()
plt.show()
```

```python
# Scatter Plot for Price VS Area Analysis by Furniture Status:

plt.figure(figsize=(8, 5))
sns.scatterplot(data = house_df, x = 'area', y = 'price', palette='viridis', hue = 'furnishingstatus')
plt.title('Price VS Area by Furniture Status', fontweight = 'bold')
plt.xlabel('Area (sq. ft)', fontweight = 'bold')
plt.ylabel('Price', fontweight = 'bold')
plt.legend(title = 'Furnishing Status', title_fontproperties = {'weight': 'bold'})
plt.grid(True)
plt.show()
```
